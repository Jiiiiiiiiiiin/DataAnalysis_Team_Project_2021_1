```python
import numpy as np
import pingouin as pg
import pandas as pd
```

```python
# Two-way mixed ANOVA
df = pg.read_dataset('mixed_anova')
aov = pg.mixed_anova(data = df, dv = 'Scores', between = 'Group',
                     within = 'Time', subject = 'Subject', correction = False, effsize = 'np2')
aov
```
